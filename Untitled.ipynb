```python
#数据清洗
import pandas as pd
import numpy as np
df = pd.read_csv('600519.csv')
def clean_stock_data(df):
    df_clean=df.copy() #数据副本
    print("原数据形状：",df_clean.shape)
    print("\n数据基本信息：")
    print(df_clean.info())
    print("\n前5行数据：")
    print(df_clean.head())
    #处理缺失值
    print(f"\n缺失值统计:")
    print(df_clean.isnull().sum())
    numeric_cols = ['open', 'high', 'low', 'close', 'pre_close', 'change', 'pct_chg', 'vol', 'amount']
    for col in numeric_cols:
        if col in df_clean.columns:
            # 价格数据用前值填充，交易量用0填充
            if col in ['vol', 'amount']:
                df_clean[col] = df_clean[col].fillna(0)
            else:
                df_clean[col] = df_clean[col].fillna(method='ffill').fillna(method='bfill')
    # 转换交易日期为datetime格式
    if 'trade_date' in df_clean.columns:
        df_clean['trade_date'] = pd.to_datetime(df_clean['trade_date'], format='%Y%m%d')
    
    # 确保数值列类型正确
    for col in numeric_cols:
        if col in df_clean.columns:
            df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
    
    #数据排序（按日期升序）
    if 'trade_date' in df_clean.columns:
        df_clean = df_clean.sort_values('trade_date').reset_index(drop=True)
    
    # 5. 去重处理
    duplicate_count = df_clean.duplicated().sum()
    print(f"\n发现 {duplicate_count} 条重复记录")
    df_clean = df_clean.drop_duplicates()
    
    # 6. 验证数据一致性
    print("\n数据一致性检查:")
    
    # 检查价格逻辑关系
    if all(col in df_clean.columns for col in ['high', 'low', 'open', 'close']):
        # 最高价应 >= 其他价格
        high_check = (df_clean['high'] >= df_clean['low']) & \
                     (df_clean['high'] >= df_clean['open']) & \
                     (df_clean['high'] >= df_clean['close'])
        
        # 最低价应 <= 其他价格
        low_check = (df_clean['low'] <= df_clean['high']) & \
                    (df_clean['low'] <= df_clean['open']) & \
                    (df_clean['low'] <= df_clean['close'])
        
        print(f"最高价逻辑正确记录: {high_check.sum()}/{len(df_clean)}")
        print(f"最低价逻辑正确记录: {low_check.sum()}/{len(df_clean)}")
        
        # 标记异常记录
        df_clean['price_anomaly'] = ~(high_check & low_check)
    
    #检测异常值（使用3σ原则）
    if 'pct_chg' in df_clean.columns:
        mean = df_clean['pct_chg'].mean()
        std = df_clean['pct_chg'].std()
        df_clean['pct_chg_anomaly'] = abs(df_clean['pct_chg'] - mean) > 3 * std
    
    #计算衍生指标
    if all(col in df_clean.columns for col in ['open', 'high', 'low', 'close']):
        df_clean['avg_price'] = (df_clean['open'] + df_clean['high'] + 
                                df_clean['low'] + df_clean['close']) / 4
        
    if all(col in df_clean.columns for col in ['high', 'low', 'pre_close']):
        df_clean['amplitude'] = (df_clean['high'] - df_clean['low']) / df_clean['pre_close'] * 100
    
    #验证涨跌幅计算
    if all(col in df_clean.columns for col in ['close', 'pre_close', 'pct_chg']):
        calculated_pct_chg = (df_clean['close'] - df_clean['pre_close']) / df_clean['pre_close'] * 100
        pct_chg_diff = abs(calculated_pct_chg - df_clean['pct_chg'])
        print(f"涨跌幅计算差异较大的记录数: {(pct_chg_diff > 0.01).sum()}")
    
    #最终数据检查
    print(f"\n清洗后数据形状: {df_clean.shape}")
    print(f"清洗后缺失值统计:")
    print(df_clean.isnull().sum())
    
    return df_clean
cleaned_data = clean_stock_data(df)
cleaned_data.to_csv('cleaned_stock_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df_clean = pd.read_csv('cleaned_stock_data.csv', encoding='utf-8-sig', parse_dates=['trade_date'], index_col='trade_date')
df_clean['daily_return'] = df_clean['close'].pct_change()
df_clean['MA5'] = df_clean['close'].rolling(window=5).mean()
df_clean['MA20'] = df_clean['close'].rolling(window=20).mean()
df_clean['volatility'] = df_clean['daily_return'].rolling(window=20).std()
df_clean['MA5_MA20_cross'] = np.where(df_clean['MA5'] > df_clean['MA20'], 1, 0)
df_clean['golden_cross'] = df_clean['MA5_MA20_cross'].diff() == 1  # 金叉信号
df_clean['dead_cross'] = df_clean['MA5_MA20_cross'].diff() == -1   # 死叉信号

# 分析金叉出现后一段时间的平均收益
golden_cross_days = df_clean[df_clean['golden_cross']]
if not golden_cross_days.empty:
    future_returns = []
    for date in golden_cross_days.index:
        future_5d_return = df_clean.loc[date:].iloc[1:6]['daily_return'].sum()  # 后5日收益
        future_returns.append(future_5d_return)
    
    avg_return = np.mean(future_returns)
    print(f"金叉信号出现后5日平均收益: {avg_return:.4f}")
```
